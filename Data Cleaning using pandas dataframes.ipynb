```python
Dealing with missing values:
```

```python
Filling missing values in SUPPLIER column with N/A in place of null values.
```

```python
df['SUPPLIER'].fillna('N/A',inplace=True)
```

```python
df.info()
```

```python
Filling missing values in ITEM TYPE column with N/A in place of null values.
```

```python
df['ITEM TYPE'].fillna('N/A',inplace=True)
```

```python
df.info()
```

```python
Filling missing values in RETAIL SALES  column with mean in place of null values.
```

```python
df['RETAIL SALES'].fillna(df['RETAIL SALES'].mean(),inplace=True)
```

```python
df.info()
```

```python
df.nunique()
```

```python
Transforming month from numbers to text.
```

```python
month_map = {1:'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',
         9:'September',10:'October',11:'November',12:'December'}
df['MONTH'] = df['MONTH'].map(month_map)
```

```python
df
```

```python
Changing case of SUPPLIER,ITEM DESCRIPTION column to Proper-case
```

```python
df['SUPPLIER'] = df['SUPPLIER'].str.title()
```

```python
df
```

```python
df['ITEM DESCRIPTION'] = df['ITEM DESCRIPTION'].str.title()
```

```python
df
```

```python
df['ITEM TYPE'] = df['ITEM TYPE'].str.title()
```

```python
df
```

```python
Deleting unwanted columns
```

```python
In some cases we dont need  to use some columns, in such scenarios we can delete unwanted columns
```

```python
df1 = df.drop(['RETAIL TRANSFERS'],axis=1)
```

```python
df1
```

```python
Deleting unwanted rows
```

```python
Categorical columns may contain uncertain values such as 'N/A','None',
```

```python
Deleting rows from the column SUPPLIER where the value is 'N/A' 
```

```python
for x in df1.index:
    if(df1.loc[x,'SUPPLIER']=='N/A'):
        df1.drop(x,inplace=True)
```

```python
df1.info()
```

```python
Deleting rows from column ITEM TYPE where value is ITEM TYPE 'N/A'
```

```python
for x in df1.index:
    if(df1.loc[x,'ITEM TYPE']=='N/A'):
        df1.drop(x,inplace=True)
```
